```python
def calculateTarget(df , i):
  target = (df.iloc[i , 0] - df.iloc[i-30 , 0])/df.iloc[i-30 , 0]
  target = target * 100
  return target
```

```python
def createIndex(df):
    values = []
    df = CloseExtractionWithNaN(df)
    df = fillNa(df)
    for i in range(30 ,df.shape[0]):
        value = calculateTarget(df , i)
        values.append(value)
    values = 30*['nan'] + values
    return values
```

```python
def Bascket_generator(Pos_indexes , Closed_indexes , turnoverList):
      
      if day == 0 :  #initialization
          Bascket_generator_day_0(Pos_indexes , turnoverList)


      else : 

          today_indexes = Pos_indexes + Closed_indexes
          prev_basket = basket[day-1]
          today_basket = list()

          print(Fore.MAGENTA + f'day {day}...is processing')
          print(Fore.GREEN + f'prev basket --->> {basket[day-1]}')
          print(Fore.YELLOW + f'today indexes can be kept in basket(because of being locked or can add)  --->> {today_indexes}')
          print(Fore.YELLOW + f'today indexes can be appended  (positives) --->> {Pos_indexes}')
          print(Fore.YELLOW + f'today indexes are locked(zeros) --->> {Closed_indexes}')



          for index in range(len(prev_basket)):
            print(Fore.WHITE + f'{prev_basket[index]} check in --->> {today_indexes}')

            if prev_basket[index] in today_indexes :
              today_basket.append(prev_basket[index]) 
              print(Fore.BLUE + f'*{prev_basket[index]}* will be kept')

            elif prev_basket[index] not in today_indexes : 
              print(Fore.RED + f'*{prev_basket[index]}* is removed')
          


          if len(today_basket) < 3 :
              free_indexes = [item for item in Pos_indexes if item not in today_basket]
              print(Fore.LIGHTYELLOW_EX + f'free indexes to choose(not in basket,are not locked) --->> {free_indexes}')

              differnt = 3 - len(today_basket)

              if len(free_indexes) >= differnt : 
                  print(Fore.WHITE +  f'sorted all indexes based on turnovers : {turnoverList}')
                  filters = filerTurnoverList(TurnoverList=turnoverList , freeIndexes=free_indexes)
                  print(Fore.WHITE +  f'filtered turnovers based on free indexes: {filters}')
                  new_orders = filters[:differnt]
              else : 
                  new_orders = random.sample(free_indexes,len(free_indexes))
              print(Fore.LIGHTYELLOW_EX + f'new orders --->> {new_orders}')
              today_basket += new_orders

          basket.append(today_basket)

          print(Fore.GREEN + f'today basket --->> {basket[day]}')
          print(Fore.MAGENTA + f'day {day}...done')
          print('-------------------------')
```

```python
def sortOfDailyTurnover(day):
    ddict = dict()
    for j,k in zip(turnoverDF.columns,turnoverDF.iloc[day , :]):
            ddict[j] = k
    sort_orders = sorted(ddict.items(), key=lambda x: x[1], reverse=True)
    return [i[0] for i in sort_orders]
```

```python
def filerTurnoverList(TurnoverList , freeIndexes):
  return [index for index in TurnoverList if index in freeIndexes ]
```
